seperating features and labels using the drop function.

```python
x=cancer.drop('TYPE',axis=1)
y=cancer['TYPE']
```

```python
y.shape
```

```python
coll=cancer.columns
for i in coll:
  print(i)
  minc=cancer[i].min()
  maxc=cancer[i].max()
  print(minc,maxc)
```

```python
from sklearn import preprocessing
min_max_scaler=preprocessing.MinMaxScaler()
x_scale=min_max_scaler.fit_transform(x)
print(x_scale)
```

Split the dataset:

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x_scale,y,test_size=0.20,random_state=0)
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(32,activation='relu',input_dim=49))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

Given:
Dataset size=349 samples


test_size=0.2 -> 20% for testing , so Training set=80% of 349 =279 samples(approx.)

batch_size=32

***steps per epoch= no.of samples/ batch size***(ceal of it)



```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,batch_size=32,epochs=15)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_pred=np.where(y_pred>0.5,1,0)
y_pred=y_pred.astype(int)
np.column_stack((y_pred,y_test))
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
